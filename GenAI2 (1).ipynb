```python
!pip install diffusers transformers torch accelerate
from diffusers import StableDiffusionPipeline
import torch
from IPython.display import display
from PIL import Image

model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id)
pipe = pipe.to("cuda")

prompt = "a futuristic cityscape at sunset with neon lights"
image = pipe(prompt).images[0]

image.save("generated_image.png")

image = Image.open("generated_image.png")
display(image)
```

```python
model_id = "stabilityai/stable-diffusion-2-1"
pipe = StableDiffusionPipeline.from_pretrained(model_id)
pipe = pipe.to("cuda")
```

```python
width = 512
height = 512
prompt = "a futuristic cityscape at sunset with neon lights"
image = pipe(prompt, width=width, height=height).images[0]
```

```python
negative_prompt = "no animals, no people"
image = pipe(prompt, negative_prompt=negative_prompt).images[0]
```

```python
image.save("custom_generated_image.png")
image = Image.open("custom_generated_image.png")
display(image)
```
